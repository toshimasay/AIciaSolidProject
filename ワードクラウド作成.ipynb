# ワードクラウド作成
1. 対象の日本語文章を単語の分かち書き配列に変換する
2. 全ての単語分かち書き配列を結合して単語毎の出現数をカウントする

```
import wordcloud
from wordcloud import WordCloud,STOPWORDS

# 単語とカウント数のdataframeを受け取ってワードクラウドを出力する関数
def make_wordcloud(df, filename):
    wordcloud = WordCloud(
        background_color="white",
        width=900, height=500,
        font_path = font_path,
        collocations=False,
    ).generate_from_frequencies(df.set_index("word").to_dict()["count"])
    
    os.makedirs(f"ワードクラウド/", exist_ok=True)
    wordcloud.to_file(f'ワードクラウド/wordcloud_{filename}.png')
```

```
# 文章を単語の分かち書き配列に変換する関数
def word_tokenize(text):
  res = requests.get("https://api-m6rhqbjcha-an.a.run.app/api/v1/parse/class/", params={"text": text})
  words = []
  for word in res.json()['result']:
    if word['basic'] != '':
      words.append(word['basic'])
    else:
      words.append(word['original'])
  return words
```

```
def get_word_lists(df, column):
  word_lists=[]
  for i, text in enumerate(df[column]):
      word_lists.append(word_tokenize(text))
  return word_lists
```

```
def get_extend_word_df(word_lists):
  extend_wordlist = []
  for wl in word_lists:
      if wl is not None:
          extend_wordlist.extend(wl)
  extend_word_df = pd.DataFrame()
  extend_word_df["word"] = extend_wordlist
  return extend_word_df
```

```
def get_count_df(word_df):
  value_counts = word_df['word'].value_counts()
  value_counts = value_counts.rename_axis('word').reset_index(name='count')
  return value_counts
```
